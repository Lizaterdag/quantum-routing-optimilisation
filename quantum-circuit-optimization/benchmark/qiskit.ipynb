```python
import matplotlib.pyplot as plt
from qiskit import QuantumCircuit, transpile
from qiskit.test.mock import FakeBoeblingen
from save_data import save_circuit
backend = FakeBoeblingen()

ghz = QuantumCircuit(4)
ghz.cx(0,1)
ghz.cx(3,2)
ghz.cx(3,0)
ghz.cx(0,2)
ghz.cx(1,2)
ghz.cx(1,0)
ghz.cx(2,3)
ghz.cx(3,0)
ghz.cx(0,2)
ghz.cx(1,3)
ghz.cx(3,2)
ghz.cx(0,1)
ghz.cx(1,2)
ghz.cx(1,0)
ghz.cx(2,3)
ghz.cx(0,2)
ghz.cx(0,1)
ghz.cx(3,2)


ghz.draw(output='mpl')
```

```python
ghz.depth()
```

```python
depths = []
for _ in range(50):

   a= (transpile(ghz,
                           backend,
                           initial_layout=[1,0,3,2], optimization_level=0
                          ).depth())
   save_circuit(a)
   depths.append(a)


plt.figure(figsize=(8, 6))
plt.hist(depths, bins=list(range(14,36)), align='left')
plt.xlabel('Depth', fontsize=14)
plt.ylabel('Counts', fontsize=14);
```
